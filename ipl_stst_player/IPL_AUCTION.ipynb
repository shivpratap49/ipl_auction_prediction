```python
x_batsmen = df_batsmen[["RUNS","MAT","INNS","4S","50"]]
y_batsmen = df_batsmen[["Price"]]
```

```python
import pickle
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_batsmen)
with open('scaled_batsmen.pkl', 'wb') as f:
    pickle.dump(scaler, f)
scaled_batsmen=scaler.transform(x_batsmen)
```

```python
scaled_batsmen
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_batsmen,y_batsmen,train_size = 0.7,random_state = 42)
```

```python
y_train
```

```python
def create_decision_tree():
    from sklearn.tree import DecisionTreeRegressor
    model = DecisionTreeRegressor()
    model.fit(x_train,y_train)
    return model
```

```python
def create_random_forest():
    from sklearn.ensemble import RandomForestRegressor
    model  = RandomForestRegressor()
    model.fit(x_train,y_train)
    return model
```

```python
def create_linearRegression():
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(x_train,y_train)
    return model
```

```python
def create_SVR():
    from sklearn.svm import SVR
    model = SVR(kernel="rbf")
    model.fit(x_train,y_train)
    return model
```

```python
def creat_KNN():
    from sklearn.neighbors import KNeighborsRegressor
    model = KNeighborsRegressor()
    
    model.fit(x_train,y_train)
    return model 
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error,root_mean_squared_error
def evaluate(model):
    x_pred=model.predict(x_train)
    
    #accuracy finding of the training data
    training_accuracy = r2_score(y_train,x_pred)
    
    #accuracy findind on test data
    y_pred = model.predict(x_test)
    
    r2 = r2_score(y_test,y_pred)
    MAE= mean_absolute_error(y_test,y_pred)
    MSE= mean_squared_error(y_test,y_pred)
    RMSE = root_mean_squared_error(y_test,y_pred)
   
    
    
    return training_accuracy,r2,MAE,MSE,RMSE
```

```python
listt = [
    ("Decision tree Regressor",create_decision_tree()),
    ("Random forest Regressor",create_random_forest()),
    ("Linear regression",create_linearRegression()),
    ("SVR",create_SVR()),
    ("KNN",creat_KNN()),
    #("Ada boost",create_adaboost())
]
```

```python
performance_data = []

for name,model in listt:
    training_accuracy,r2,MAE,MSE,RMSE = evaluate(model)
    performance_data.append([
      name,training_accuracy,r2,MAE,MSE,RMSE
    ])
    
performance_dataframe = pd.DataFrame(performance_data,columns = ["Model","training_accuracy","r2","MAE","MSE","RMSE"])
performance_dataframe 
```
